```python
from bs4 import BeautifulSoup
from splinter import Browser
import requests
```

```python
executable_path = {'executable_path': '/usr/local/bin/chromedriver'}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
url = 'https://mars.nasa.gov/news/'
browser.visit(url)
```

```python
html = browser.html
soup = BeautifulSoup(html, 'html.parser')
```

```python
iterable_list=soup.find('li',class_='slide')
print(iterable_list)
```

```python
news = iterable_list.find('div', class_='content_title')
# print(news)
# a=news.find('a')
news.text
```

```python
# collect the latest News Title and Paragraph Text.
# Assign the text to variables that you can reference later
# Example:
# news_title = "NASA's Next Mars Mission to Investigate Interior of Red Planet"

# news_p = "Preparation of NASA's next spacecraft to Mars, InSight, has ramped up this summer,
# on course for launch next May from Vandenberg Air Force Base in central California 
# -- the first interplanetary launch in history from America's West Coast."
titles=[]
paragraph=[]
for x in range(1, 6):
    html = browser.html
    soup = BeautifulSoup(html, 'html.parser')

    news = soup.find_all('div', class_='content_title')

    for new in news:
        if (new.a):
            if (new.a.text):
                titles.append(new)
                
    paragprahs=soup.find_all('div', class_='article_teaser_body')
    for row in paragprahs:
        paragraph.append(row)
#     browser.links.find_by_partial_text('Next')
```

```python
for x in range(len(titles)):
    print('-------------------------------------------------------------------------------------------')
    print(f" news_title_{x}: {titles[x].text}")
    print(f" new_p{x}: {paragraph[x].text}")
    print(url+titles[x].a['href'])
```

```python
# close the browser
browser.quit()
```
